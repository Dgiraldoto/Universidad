```python
from zmq.constants import NULL
import json
import requests
Information=[]
from pprint import pprint
response = requests.get(
  "https://fp.josedomingo.org/lmgs/u05/ej2.json")
Information=json.loads(response.text)
#pprint (Information)

def n_tareas(): # ==> Ejercicio 1
  cont=0
  for i in Information:
    cont+=1
  return cont

def pruebas_2h(): # ==> Ejercicio 2
  titulos=[]
  titulo=""
  for i in Information:
    for j in i:

      if(j=="Titulo"): 
        titulo=i[j]
        #print(titulo)
      
      if(j=="Horas"):
        if(i[j]>2):
          titulos.append(titulo)
  return titulos

def ej_3(): # ==> Ejercicio 3
  for i in Information:
    tipo={}
    if((i["TipoFormacion"]=='NoPresencial')|(i["TipoFormacion"]=='No Presencial')):
      tipo[i["TipoFormacion"]]=i['Profesorado'][0]['NombreCompleto']
  return tipo


def codigoProfesores(): # ==> Ejercicio 4
  id=input("Ingrese el Id de la prueba: ")
  dic={}
  prof=[]
  for i in Information:
    if(i.get("$id")==id):
      datos_prof=i.get("Profesorado")
      titulo=i.get("Titulo")
      for j in datos_prof:
        teach=j.get("NombreCompleto")
        prof.append(teach)

      dic[titulo]=prof
  print(dic)


def archivo_json(): # ==> Ejercicio 5
  archivo=[]
  dic={}
  for i in Information:
    dic[i["Titulo"]]=i["Profesorado"]
  archivo.append(dic)
  with open('/content/drive/MyDrive/archivo.json','w') as f:
    json.dump(archivo, f)
  f.close
  return archivo


print("Ej1: ")
print(n_tareas())
print("Ej2: ")
print(pruebas_2h())
print("Ej3: ")
print(ej_3())
print("Ej4: ")
codigoProfesores()
print("Ej5: ")
archivo_json()
```
