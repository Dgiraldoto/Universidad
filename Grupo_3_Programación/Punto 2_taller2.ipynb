```python
'''Escribir un algoritmo que pida un número y tambien un porcentaje
entre 0 y 100 que puede ser decimal, el algoritmo debe entregar el
valor del procentaje del primer número. Si el usuario ingresa un dato
fuera del rango (0 a 100), se debe mostrar un error y volver a pedir el
porcentaje'''
a=int(input("Digite un numero cualquiera\n"))#definimos 'a' como entera y se le asigna la primera entrada
b=int(input("Digite el porcentaje a calcular (0 a 100)\n"))#definimos 'b' como entera y se le asigna la segundo entrada
if(b>100 or b<0):#iniciamos una condicion. Sí 'b' es menor a 0 ó mayor a 100, esta entra en el 'if'
  while(b>100 or b<0):#Se crea un bucle, donde la condicion para este, es que 'b' sea menor a 0 o mayor a 100
    b=int(input("Digite un numero entre 0 y 100\n"))#Se pide nuevamente la entrada numero 2, restringiendole al usuario un valor entre 0 y 100
print(f"El {b}% de {a} es {round((a)/(b/100),3)}")#se le muestra al usuario el resultado obtenido a partir de la entreda 'a' y 'b'
```
